```python
 
# sanitizing the data
# removing nulls and redundant columns
df = df.dropna()
df = df.drop(['TRIP_START', 'TRIP_END'], axis=1) 

# using the start and and cordinates to generate the distances covered by each
# since the coordinates are coloumn vectors, i'm doing the harvesine to do a 
# vector evaluation (speed eficiency). The results are added to the dataset in 
# the 'distance_km' column 
lat_rad = np.radians(df['LOCATION_LAT_START'])
lon_rad = np.radians(df['LOCATION_LONG_START'])
long_diff = lon_rad - np.radians(df['LOCATION_LONG_END'])
lat_diff = lat_rad - np.radians(df['LOCATION_LAT_END'])
df['distance_km'] = 6367 * 2 * np.arcsin(np.sqrt(np.sin(lat_diff/2)**2 + np.cos(np.radians(df['LOCATION_LAT_END'])) * np.cos(lat_rad) * np.sin(long_diff/2)**2))

df.head(5)
```

# Building and Training a classifier

I'm using the random forest classifier -- coulda used a decision tree (Random forest is better than decision tree as if uses more features.) , k-nearest neighbours (similar to clustering, might be used for recommendor systems) or cnn (will be used later) 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(
                                      X, y, 
                                      test_size=0.2,        # split ratio
                                      random_state=118833   # randomisation seed
                                   )
```

```python
# classifier and hyperparameters
classifier = RandomForestClassifier(random_state=50, n_jobs=-1)
params = {
    'max_depth': [2,3,5,10,20], 
    'min_samples_leaf': [3,5,10,20,50],
    'n_estimators': [10,25,30,50,100]
} # ,3,5,10,20 , 3, 5,10,20,50,25,30,50,100
```

```python
# feeding the classifier to a gridsearch algorithm for hyperparameter tuning
# This ensures a best model is selected for the given list of hyperparameters
# I'm choosing the 'accuracy' metric for the performance tuning. 
grid_search = GridSearchCV(
    estimator=classifier,
    param_grid=params,
    cv = 4,
    n_jobs=-1, 
    verbose=1, 
    scoring="accuracy"
    
)
```

```python

%%time
grid_search.fit(X, y) # training the model
```

```python
# our model is the best estimator seleted by the grid search
model = grid_search.best_estimator_
model
```

# Model Evaluation

```python
# feature importances: shows which features contribute more principal
# components to the model
feature_importance = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": model.feature_importances_
})

feature_importance.sort_values(by='Importance', ascending=False)
```

```python
# Visualize the model
from sklearn.tree import plot_tree

plt.figure(figsize=(180,140), dpi=120)
plot_tree(model.estimators_[8], feature_names = X.columns,class_names=['ICE', "EV", "Hybrid"],filled=True);
```

```python
# check the score of the metric specified in the grid search, 
# 'accuracy' in this case
score = model.score(X_test, y_test)
score # model has an accuracy around 65%
```
